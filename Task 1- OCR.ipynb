## Import some modules

### Handwitten Image to Text Converter

```python
# An image proccessing library
import cv2 
```

```python
# an image to text library
import pytesseract 
```

```python
# used for mathematics but can be used in image proccessing
import numpy as np 
```

```python
# display an image
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
# Configure the module
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\NARSAIAH\AppData\Local\tesseract.exe'
```

```python
# Make the image grey
# opening an image from the source path
img = cv2.imread(r'C:\Users\NARSAIAH\dataimages\alphabets.jpg')
plt.imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gray, img_bin = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
gray = cv2.bitwise_not(img_bin)
kernel = np.ones((2, 1), np.uint8)
img = cv2.erode(gray, kernel, iterations=1)
img = cv2.dilate(img, kernel, iterations=1)
```

```python
# Use OCR to read the text from the image
out_below = pytesseract.image_to_string(img)
# Print the text
print(out_below)
```

```python
# write text in a text file and save it to source path   
with open('handwritten2text.txt', mode ='w') as file:
    file.write(out_below)
    print('Done')
```

### Printed Image to Text converter

```python
# Import some modules
import cv2 
# An image proccessing library
import pytesseract
# an image to text library
import numpy as np 
# used for mathematics but can be used in image proccessing
import matplotlib.pyplot as plt
import matplotlib.image as img

# Configure the module
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\NARSAIAH\AppData\Local\tesseract.exe'

# Make the image grey
img = cv2.imread(r'C:\Users\NARSAIAH\example.png')
plt.imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
gray, img_bin = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
gray = cv2.bitwise_not(img_bin)
kernel = np.ones((2, 1), np.uint8)
img = cv2.erode(gray, kernel, iterations=1)
img = cv2.dilate(img, kernel, iterations=1)

# converts the image to result and saves it into result variable
out_below = pytesseract.image_to_string(img)

# Print the text
print(out_below)

# write text in a text file and save it to source path   
with open('printed2text.txt', mode ='w') as file:
    file.write(out_below)
    print('Done')
```
